```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Step 2: Load one CSV file (you can loop through later)
DATA_PATH = os.path.join("..", "data", "raw")   # ".." to go up one folder
file_name = r"D:\Downloads\csvDataset\Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv"  # change this to any CSV you want to explore
file_path = os.path.join(DATA_PATH, file_name)

df = pd.read_csv(file_path)
```

```python

# Step 3: Inspect the dataset
print("Dataset shape:", df.shape)
print("\n--- Columns ---")
print(df.columns.tolist())
print("\n--- First 5 Rows ---")
display(df.head())
```

```python
# Step 4: Check datatypes and missing values
print("\n--- Info ---")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Step 5: Explore label distribution
if "Label" in df.columns:
    print("\n--- Class Distribution ---")
    print(df["Label"].value_counts())

    # Plot label distribution
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x="Label", order=df["Label"].value_counts().index)
    plt.xticks(rotation=45)
    plt.title("Attack Type Distribution")
    plt.show()
```

```python

# Step 6: Explore some numerical features
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[num_cols].describe().T
```

```python

# Example histogram for Flow Duration (if exists)
if "Flow Duration" in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df["Flow Duration"], bins=50, kde=True)
    plt.title("Flow Duration Distribution")
    plt.show()
```
